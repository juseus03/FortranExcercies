```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load de data from the file and extract the arrays created in RK4.f90
data=np.loadtxt("results.txt")
x=data[0]
y=data[1]
```

```python
#Plot the results of RK4 and the theoric function
plt.plot(x,y, 'ko')
plt.plot(x,np.exp(-x))
plt.xlabel('x')
plt.ylabel('y(x)')
```

```python
#Plot the error 
plt.plot(x,abs(np.exp(-x)-y), 'ko')
plt.xlabel('$x$', size=30)
plt.ylabel('$|y_{\mathrm{true}}-y_{\mathrm{euler}}|$',size=30)
```
